```python
#loading the requried libraries
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
marvel=pd.read_csv('marvel.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment_x'].value_counts()
```

```python
#Good_characters
marvel['Alignment_x']=='good'
```

```python
good=marvel[marvel['Alignment_x']=='good']
good
```

```python
good.sort_values(by=['Speed'],ascending=False).head()
```

```python
good.sort_values(by=['Power'],ascending=False).head()
```

```python
good_max_power=good[good['Power']==100]
```

```python
good_max_power
```

```python
good_max_power.shape
```

```python
good_max_power.sort_values(by=['Total'],ascending=False).head()
```

```python
plt.figure(figsize=(6,5))
plt.bar(list(good_max_power['Name'])[0:5],list(good_max_power['Total'])[0:5],color='g')
plt.show()
```

```python
#Bad_characters
bad=marvel[marvel['Alignment_x']=='bad']
bad.head()
```

```python
#Making a histogram for speed for good superheroes
plt.figure(figsize=(8,5))
plt.hist(good['Speed'])
plt.title('Distribution of Speed')
plt.xlabel('Speed')
plt.show()
```

```python
#making of histogram for combat for supervillains
plt.figure(figsize=(8,7))
plt.hist(bad['Combat'])
plt.title('Distribution of combat')
plt.xlabel('combat')
plt.show()
```
